### EXPLORATORY DATA ANALYSIS ON WORLD HAPPINESS INDEX DATASET

### ABOUT THE DATASET




The World Happiness Report is a landmark survey of the state of global happiness . This study analyzes the association between the Happiness Index Score in 2023 and a set of independent variables such as 'Overall rank', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity' and 'Perceptions of corruption'. The objective of this study is to investigate the impact of these independent variables on the level of happiness of individuals during this year.

#### Overview of the Data


"Overall rank: List of ranks of different countries from 1 to 156
Country or region: List of the names of different countries.
Score: List of happiness scores of different countries.
GDP per capita: The GDP per capita score of different countries.
Social support: The social support of different countries.
Healthy life expectancy: The healthy life expectancy of different countries.
Freedom to make life choices: The score of perception of freedom of different countries.
Generosity: Generosity (the quality and well_being behaviour of individuals score of different countries.
Perceptions of corruption: The score of the perception of corruption in different countries.

   By examining the relationship between these independent variables and the Happiness Index Score, the study aims to contribute to the understanding of the factors that impact individuals' happiness levels"

#### Keywords

 Happiness Index Score, GDP per capita, social support, healthy life expectancy, freedom to make life choices, perceptions of corruption.
